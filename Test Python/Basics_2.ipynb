```python
def read_csv(csv_name):
    arquivo = open(csv_name,"r")
    data_string = arquivo.read()
    rows = data_string.split("\n")
    string_list = rows[1:len(rows)-1] # removendo header
    final_list = []
    int_fields = []
    for row in string_list:
        string_fields = (int, row.split(","))
        int_fields.append(string_fields)
    final_list.append(int_fields)
    return final_list
```

```python
cdc_list = read_csv("US_births_1994-2003_CDC_NCHS.csv")
print (cdc_list)
```
